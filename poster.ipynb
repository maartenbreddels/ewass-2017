The code below is for the 3d selection (only works in a live Jupyter notebook)

```python
import ipyvolume.pylab as p3
```

```python
import vaex
hdz = vaex.datasets.helmi_de_zeeuw.fetch() # this will download it on the fly
hdz.plot([["x", "y"], ["Lz", "E"]], f="log", figsize=(12,5), show=True);
```

```python
hdz.plot_widget("Lz", "E", f='log')
hdz.plot_widget("x", "y", f='log')
```

```python
w3d = hdz.plot_widget("x", "y", "z", vx='vx', vy='vy', vz='vz', backend='ipyvolume', f='log',shape=100)
```

```python
w3d.vcount_limits = [60, 10000]
```
